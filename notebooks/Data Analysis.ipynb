# Data Analysis with Pandas and sklearn
For this tutorial we'll be using the Iris dataset from sklearn.

You will first need to install sklearn before using this dataset:

`pip install sklearn`  
or  
`conda install sklearn`

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()

iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df["class"] = iris.target
iris_df.head()
```

## Summary Statistics
We can get a quick glance at the distributions of our data using pandas describe() method: 

```python
iris_df.describe()
```

We can also drill down to a single column or subset of columns: 

```python
iris_df["class"].describe()
```

### How many rows do we have for each class? 

```python
iris_df["class"].value_counts()
```

### Pandas Profiling
We can get more detailed statistics using pandas profiling module. 

First, from the command line, install dependencies:

`pip install pandas`  
`pip install matplotlib`  
`pip install pandas-profiling` 

```python
import pandas_profiling
# ignore irritating pandas warnings
import warnings
warnings.filterwarnings('ignore')
pandas_profiling.ProfileReport(iris_df)
```

## Pre-processing
Pandas makes it very easy to pre-process your data. Here are some common things to look for. 

### Check Data Types
Make sure your data types are what you think they are: 

```python
iris_df.dtypes
```

### Plotting Frequency

```python
iris_df["class"].value_counts().plot(kind="bar")
```

### Check for null values
If our dataframe contains null values, we have to decide how to deal with them, depending on the dataset: 
 
- ignore missing values
- impute missing values
- remove missing values

You'll learn more about how to deal with these issues in class, but here's how to find them: 
    

```python
iris_df.isnull().values.any()
```

### Filtering 
Pandas is also great for SQL filtering of datasets. Let's say we want to find out if any of our irises have a sepal length greater than 7: 

```python
iris_df[iris_df["sepal length (cm)"] > 7]
```
